So we want to look at all the columns that have the same 'Journal title.' After collecting the costs for each entry, if any of the entries are above the specific journal title's 95th percentile, change it to that 95th percentile (Winsorizing).

```python
plos_one = (df.lc_journal_title == 'plos one')
```

```python

plos_one_Cost = df[plos_one]['Cost']
plos_one_Cost
```

```python
np.percentile(plos_one_Cost, 95)
```

```python
plos_one_winsor_cost = scipy.stats.mstats.winsorize(plos_one_Cost, limits = 0.05)
plos_one_winsor_cost
```
